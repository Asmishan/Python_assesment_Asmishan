Scenario

```python
def check_bonus(attendance, late_count):    #Function is defined with 2 parameters
  if attendance >= 90 and late_count <= 2:  #Checking the condition
    return True                             #Checking the condition
  else:
    return False
attendance = float(input("Attendance Percentage: "))     #Taking inputs from the user
late_count = int(input("Late Arrivals: "))

if check_bonus(attendance, late_count):                  #Value is passed
  print("Eligible for Bonus")
else:
  print("Not Eligible for Bonus")

# TODO 3:
# Call the function

# TODO 4:
# Print
# Eligible for Bonus
# OR
# Not Eligible for Bonus



# Sample Input
# 94
# 1
# Output
# Eligible for Bonus
```

The rating is converted into feedback labels.

5	Excellent
4	Good
3	Average

The program should process multiple customer records and display the final feedback summary.



```python
def get_feedback(rating):                          #Function is defined for checking the rating
    if rating >= 5:                                #Checking for various conditions
        return "Excellent"
    elif rating == 4:
        return "Good"
    elif rating == 3:
        return "Average"
    else:
        return "Poor"


def feedback_summary():                             #Function for summary
    n = int(input("Enter number of customers: "))
    feedback = {}                                   #Empty dictionary is created
    for cus in range(n):
        customer = input("Enter name:-")            #Taking inputs from the user
        rating = int(input("Enter rating:-"))
        feedback[customer] = get_feedback(rating)   #Function call and adding name,rating to the dict
    print("\nCustomer Feedback Summary")
    for customer, feedback in feedback.items():
        print(customer, ":", feedback)              #For display

feedback_summary()


 # TODO 1
    # Return feedback
# TODO 2
        # Find feedback

        # TODO 3
        # Store customer and feedback
        # TODO 4
    # Display all customer feedback

# Sample Input
# 3
# Alice
# 5
# John
# 3
# Maria
# 2
# Output
# Customer Feedback Summary

# Alice : Excellent
# John : Average
# Maria : Poor
```

Sales Performance Dashboard
Scenario

A retail company wants to generate a quick sales summary.

Each transaction contains:

Salesperson SalesAmount

Example

Asha 15000
Rahul 22000
Asha 8000
John 12000
Rahul 6000

The program should

Calculate total company sales.
Calculate total sales for each salesperson.
Identify the top performer.
Display the sales summary.

```python
def sales_dashboard():                                #Function is defined
    n = int(input("Enter number of sales records: ")) #Taking input from user
    sales = {}                                  #Empty dict
    total_sales = 0
    top_salesperson = ""
    highest_sale = 0
    for i in range(n):
      name=input("Enter name:-")                #Reading input from the user
      amount=input("Enter amount:-")
      amount = float(amount)
      if name in sales:                         #For calculating the sale
            sales[name] += amount
      else:
            sales[name] = amount
      for name, amount in sales.items():        #Calculating highest sale and salesperson
        if amount > highest_sale:
            top_salesperson = name
            highest_sale = amount
        elif amount == highest_sale:
            top_salesperson = name
            highest_sale = amount

        total_sales = sum(sales.values())         #Total sale calculation

    print("\nTotal Company Sales:", total_sales)  #Display of total sale, top performer and summary
    print("Top Performer:", top_salesperson)
    print("\nSales Summary")
    for name, amount in sales.items():
        print(name, ":", amount)
sales_dashboard()





# TODO 1
        # Update salesperson sales

    # TODO 2
    # Calculate total company sales

            # TODO 2
    # Calculate total company sales

    # TODO 3
    # Identify top performer
    # TODO 4
    # Display salesperson-wise sales



# Sample Input
# 5
# Asha 15000
# Rahul 22000
# Asha 8000
# John 12000
# Rahul 6000
# Output
# Total Company Sales: 63000

# Top Performer: Rahul

# Sales Summary

# Asha : 23000
# Rahul : 28000
# John : 12000
```
